```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
print(x)
print(x.ndim)
```

```python
# 2D array (matrix)
multi_arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(multi_arr_2d)
# Output:
# [[1 2 3]
#  [4 5 6]]

# 3D array
multi_arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(multi_arr_3d)
```

```python
# 3x4 array of zeros
zeros_arr = np.zeros((3, 4))
print(zeros_arr)

# 2x2 array of ones
ones_arr = np.ones((2, 2))
print(ones_arr)
```

```python
print(f"2D Array dimensions (ndim): {multi_arr_2d.ndim}") # Output: 2
print(f"2D Array shape: {multi_arr_2d.shape}")             # Output: (2, 3)

print(f"3D Array dimensions (ndim): {multi_arr_3d.ndim}") # Output: 3
print(f"3D Array shape: {multi_arr_3d.shape}")             # Output: (2, 2, 2)
```

```python
rng = np.random.default_rng()

# 2x3 array with random floats between 0.0 and 1.0
random_arr_2d = rng.random((2, 3))
print(random_arr_2d)

# 3x3 array with random integers between a low (inclusive) and high (exclusive) value
random_int_arr = rng.integers(low=0, high=10, size=(3, 3))
print(random_int_arr)
```

```python
# Generate random array
```

```python
# 'rand' generates random value between 0 and 1
x=np.random.rand(4)
print(x)
print()
x=np.random.rand(4,5)
print(x)
```

```python
# 'randn' generates value close to 0 (positive/negativede)
x=np.random.randn(3)
print(x)
print()
x=np.random.randn(2,4)
print(x)
```

```python
# 'ranf' generates random number [0.0, 1.0)
x=np.random.ranf(3)
print(x)
print()
x=np.random.ranf((2,4))
print(x)
```

```python
# 'randint' generates random number between given 2 numbers
x=np.random.randint(3,15,4)
print(x)
```

```python
# Data Type
x=np.array([1,2,3,4], dtype=np.int8)
print("Data Type", x.dtype)
print()
x=np.array([1,2,3,4,"A","B","C"])
print("Data Type", x.dtype)
```

```python
# Numpy arithmatic operations
x=np.array([1,2,3,4])
add=x+4
print(add)

x1=np.array([1,2,3,4])
x2=np.array([4,3,6,7])
add1=x1+x2
print(add1)
```

```python
x1=np.array([1,2,3,4])
x2=np.array([5,6,7,3])
print(np.add(x1, x2))
print(np.subtract(x1, x2))
print(np.multiply(x1, x2))
print(np.divide(x1, x2))
print(np.mod(x1, x2))
```

```python
x1=np.array([[1,2,3,4], [5,6,7,8]])
x2=np.array([[5,6,7,3], [3,2,0,8]])
print(np.add(x1, x2))
```

```python
x=np.array([1,2,3,4,5])
print(np.min(x))
print(np.max(x))
```

```python
x=np.array([1,2,3,4,5,6])
print(x.reshape(3,2))
```

```python
x=np.array([[1,2,3]])
y=np.array([[1],[2],[3]])
print(x+y)
```

```python
x=np.array([[1,2,3],[4,5,6]])
print(x.shape)
y=np.array([[1,2],[4,5],[7,8]])
print(y.shape)
print(x*y)
```
