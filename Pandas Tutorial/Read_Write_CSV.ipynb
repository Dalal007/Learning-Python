# Write CSV

```python
import pandas as pd
```

```python
dict={"a":[1,2,3,4,5],"s":[6,7,8,9,10],"d":[11,12,13,14,15]}
var=pd.DataFrame(dict)
var
```

```python
var.to_csv("demo.csv", index=False, header=['p','q','r'])
var
```

# Read CSV

```python
csv=pd.read_csv("customers-100.csv")
csv
```

```python
# Get number of rows with specific columns
# csv=pd.read_csv("customers-100.csv", nrows=2, usecols=["Index", "First Name", "City"])
csv=pd.read_csv("customers-100.csv", nrows=2, usecols=[1, 3, 6])
print(type(csv))
csv
```

```python
# Skip rows, index_col
csv=pd.read_csv("customers-100.csv",skiprows=[1], index_col="Customer Id")
csv
```

```python
# skip header line
csv=pd.read_csv("customers-100.csv",header=1)
csv
```

```python
# column names
csv=pd.read_csv("customers-100.csv",names=["col1","col2","col3","col4","col5","col6","col7","col8","col9","col10","col11","col12"])
csv
```
