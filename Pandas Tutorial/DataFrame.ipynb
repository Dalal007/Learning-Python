# DataFrame

```python
import pandas as pd
```

```python
l=[1,2.3,4,5,6]
var=pd.DataFrame(l)
print(var)
print(type(var))
```

```python
d={"a":[1,2,3,4,5], "b":[6,7,8,9,10], "c":[11,12,13,14,15]}
var=pd.DataFrame(d)
print(var)

# To get specific columns

var1=pd.DataFrame(d, columns=["a", "c"], index=['a','s','d','f','g'])
print(var1)

print(var["b"][4])
```

```python
lst=[[1,2,3,4], [5,6,7,8]]
var=pd.DataFrame(lst)
print(var)
```

```python
sr={"a":pd.Series([1,2,3,4]), "b":pd.Series([5,6,7,8])}
var=pd.DataFrame(sr)
print(var)
```
