# dropna & fillna

```python
import pandas as pd
```

```python
csv=pd.read_csv("sample.csv")
csv
```

```python
# drop by rows
csv.dropna()
```

```python
# subset: remove entire row if given column has null value
csv.dropna(subset=["CHANGE"])
```

```python
var=pd.read_csv("sample.csv")
var.fillna("ASDASD")
```

```python
var.fillna({"TODAY-VOLUME":"asdasd","AVG.VOLUME":"zxczxc","CHANGE":"rtyrty"})
```
