# Merge

```python
import pandas as pd
```

```python
var1=pd.DataFrame({"A":[1,2,3,4], "B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,5], "C":[21,22,23,24]})
pd.merge(var1,var2)
```

```python
# how to merge: left, right, inner, outer
# pd.merge(var1,var2, how="left")
# pd.merge(var1,var2, how="right")
# pd.merge(var1,var2, how="inner")
pd.merge(var1,var2, how="outer")
```

# Concat

```python
var1=pd.Series([1,2,3,4])
var2=pd.Series([10,20,30,40])
pd.concat([var1,var2])
```
