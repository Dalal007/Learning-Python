# Join

```python
import pandas as pd
```

```python
v1=pd.DataFrame({"A":[1,2,3,4], "B":[5,6,7,8]})
v2=pd.DataFrame({"C":[11,12,13], "D":[100,200,300]})

# v1.join(v2)
v2.join(v1, how="outer") # how: left, right, inner, outer
```

# Append

```python
v1=pd.DataFrame({"A":[1,2,3,4], "B":[5,6,7,8]})
v2=pd.DataFrame({"C":[11,12,13,14], "D":[100,200,300,400]})

v1.append(v2)
```
