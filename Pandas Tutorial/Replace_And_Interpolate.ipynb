# Interpolate

```python
var=pd.read_csv("sample.csv")
var
```

```python
# List the columns that contain numbers
numeric_cols = ['TODAY-VOLUME', 'AVG.VOLUME', 'CHANGE']

# Interpolate only those columns
var[numeric_cols] = var[numeric_cols].interpolate()
var
```
