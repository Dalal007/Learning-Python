# Series

```python
import pandas as pd
```

```python
x=[3,4,5,6,7,8,9]
var=pd.Series(x, name="Name of the Series", dtype="int8", index=['a','b','c','d','e','f','g'])
print(var)
print(type(var))
print(var['f'])
```

```python
dict = {"name": "Yash Dalal", "age": 27, "address": "Ahmedabad"}
var=pd.Series(dict)
print(var)
print(type(var))
```

```python
s1=pd.Series(10, index=[1,2,3,4,5])
s2=pd.Series(5, index=[1,2,3])
print(s1+s2)
```
