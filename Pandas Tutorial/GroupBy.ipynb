# GroupBy

```python
import pandas as pd
```

```python
var=pd.DataFrame({"name":['a','s','d','f','z','x','c','a','s','d','v'], "s1":[10,11,15,45,3,7,51,54,56,23,49], "s2":[33,15,6,48,75,78,24,65,95,13,57]})
var
```

```python
var_new=var.groupby("name")
var_new
```

```python
for x,y in var_new:
    print(x)
    print(y)
    print()
```

```python
# To get specific grouped key data
var_new.get_group("a")
```

```python
# Get min and max values from group
# var_new.min()
var_new.max()
```

```python
# Convert to the list
li=list(var_new)
li
```
