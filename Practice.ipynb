```python
v=np.array([1,5,2,4,5,8,9,1,4,5])
v1=np.array([3,6,3,5])
print(v)
print(np.concatenate([v,v1]))
print(np.unique(v))
print(np.sort(v1))
```

```python
v1=np.random.randint(1,5,size=(3,4))
v2=np.random.randint(1,6,size=(3,4))
print(v1)
print(v2)
print(np.vstack([v1,v2]))
print(np.hstack([v1,v2]))
print(np.hstack([v1,v2]).flatten())
```
