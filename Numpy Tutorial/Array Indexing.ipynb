```python
import numpy as np
```

#### Numpy indexing and slicing of vectors

```python
arr=np.arange(1,21)
arr
```

```python
arr[5]
```

```python
arr[4:11]
```

```python
arr[15:5:-1]
```

#### Numpy indexing and slicing of matrix

```python
arr=np.arange(1,31).reshape(5,6)
arr
```

```python
#Get specific row
arr[1]
```

```python
# Get specific element
arr[3,4]
```

```python
# Get specific slices
arr[1:4,2:4]
```

```python
# Get specific column
arr[:,2]
```

#### Boolean indexing

```python
arr=np.arange(1,21)
arr
```

```python
bool_index=arr%2==0
bool_index
```

```python
arr[bool_index]
```
