```python
import numpy as np
```

### Array Operations

#### Arithmatic operations

```python
a1=np.array([1,2,3,4,5])
a2=np.array([6,7,8,9,10])
```

```python
print(a1+a2)
print(a2-a1)
print(a1*a2)
print(a2/a1)
print(a2%a1)
```

#### Broadcasting

```python
l=[1,2,3,4,5]
arr=np.array(l)
arr+5
```

```python
arr2=np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2+10
```

#### Deep and shallow copy

```python
# Shallow Copy
a=np.arange(1,11)
b=a
b[0]=11
print(b)
print(a)
```

```python
# Deep Copy
arr=np.arange(1,11)
slice=arr[0:5].copy()
slice*=10
print(slice)
print(arr)
```

#### Matrix operations

```python
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
# multiplication(dot product)
print(a@b)
# Transpose
print(a.T)
print(a+b)
print(a-b)
```

### Advance array manipulation

#### Stacking array

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

#### Spliting array

```python
arr=np.arange(1,17).reshape(4,4)
arr
```

```python
# horizontal split
np.hsplit(arr,2)
```

```python
# vertical split
np.vsplit(arr,2)
```
