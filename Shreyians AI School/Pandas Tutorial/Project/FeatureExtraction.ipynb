```python
import numpy as np
import pandas as pd
```

#### loading data

```python
df=pd.read_csv("anime.csv")
df.head()
```

#### make a new column for episode count

```python
df.loc[1]["Title"]
```

```python
def extract_episodes(txt):
     return int((txt.split("(")[1]).split(" ")[0])
```

```python
df["Episodes"]=df["Title"].apply(extract_episodes)
df.head()
```

#### make a new column for timestamp

```python
def extract_timestamp(txt):
    return (txt.split(" - ")[0][-1:-9:-1][::-1]+" - "+txt.split(" - ")[1][0:8])
```

```python
df["Total Time"]=df["Title"].apply(extract_timestamp)
df.head()
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1 # +1 to include the starting month
    except:
        return None

df["Months"]=df["Total Time"].apply(calculate_total_months)
df.dropna(inplace=True)
df
```

#### which anime has a highest score

```python
df[df["Score"]==df["Score"].max()]
```

#### give me top 5 highest scoring anime

```python
df[df["Score"]==df["Score"].sort_values(ascending=False)].head(5)
```

#### which anime has highest episode count

```python
df[df["Episodes"]==df["Episodes"].max()]
```

#### animes with top 5 episode counts

```python
df.sort_values(by="Episodes", ascending=False).head()
```

#### which is the longest running anime

```python
df[df["Months"]==df["Months"].max()]
```
