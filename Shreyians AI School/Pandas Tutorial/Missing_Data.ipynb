### Finding Missing Data

```python
import numpy as np
import pandas as pd
```

```python
data={
    "A":[1,2,np.nan,4,5],
    "B":[1,2,3,4,5],
    "C":[1,2,3,np.nan,np.nan],
    "D":[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
df
```

```python
# To check null values
df.isna()
```

```python
# To check how many null values a columns has
df.isna().sum()
```

```python
# To check any kind of missing values present in each column
df.isna().any()
```

### Removing Missing Data

```python
df.dropna()
```

```python
df.dropna(thresh=2)
```

### Filling the Missing Data

```python
df.fillna(0)
```

```python
# To replace values column wise
values={"A":0,"B":100,"C":300,"D":400}
df.fillna(value=values)
```
