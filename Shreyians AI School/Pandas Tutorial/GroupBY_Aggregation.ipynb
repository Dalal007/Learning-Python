```python
import numpy as np
import pandas as pd
```

# GroupBy

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
df
```

```python
# Groupby the category and calculate the sum of sales
sales_sum=df.groupby("Category")["Sales"].sum()
sales_sum
```

```python
# Groupby store and calculate the sum of sales
sales_sum=df.groupby("Store")["Sales"].sum()
sales_sum
```

```python
# Groupby multiple columns
# Groupby category and store
sales_sum=df.groupby(["Category","Store"])["Sales"].sum()
sales_sum
```
