# Valid Sudoku

```python
import numpy as np
```

```python
sudoku = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
b=np.sum(sudoku,axis=1)
print(b==45)

c=np.sum(sudoku,axis=0)
print(c==45)
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        sum=(sudoku[i:i+3,j:j+3]).sum()
        if sum!=45:
            print("Invalid sudoku")
            break
```

```python
# Columns = [Age, Maths, Science]
data=np.array([[18,85,78],[19,92,88],[17,76,95],[18,65,70],[20,90,85]])
```

```python
# Get the shape of the matrix
data.shape
```

```python
# Find the average age of students
data[:,0].mean()
```

```python
# Extract Maths marks of all students
data[:,1]
```

```python
# Find the highest Science marks
data[:,2].max()
```

```python
# Get the details of students who scored more than 90 in Maths
maths_highest=data[:,1]>90
data[maths_highest]
```

```python
# Increase Maths marks of all students by 5
maths_marks=data[:,1]+5
maths_marks
```

```python
# Find how many students are younger than 19
under19=data[:,0]<19
data[under19]
```

```python
# Calculate avg marks in each subject(column-wise mean)
avg_subject=data[:,1:].mean(axis=0)
avg_subject
```

```python
# Get data of students who scored atleast 80 in both subjects
more_than_80=((data[:,1]>=80) & (data[:,2]>=80))
data[more_than_80]
```

```python
# Replace all Science marks < 75 with 0
data[:,2][data[:,2]<75]=0
data
```
