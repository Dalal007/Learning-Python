```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Matplotlib Basics

```python
x=np.linspace(1,5,10)
y=x**2
```

```python
plt.plot(x,y)
plt.title("Number to square")
plt.xlabel("Number")
plt.ylabel("Square")
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(y,x)

plt.subplot(2,2,3)

plt.subplot(2,2,4)
```

### Object Oriented Way

```python
fig=plt.figure(figsize=(10,6), dpi=150) # figsize: (width, height), dpi: image quality
axis1=fig.add_axes([0.1,0.1,1,1]) # [left spacing, bottom spacing, width, height]
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.5,0.3,0.3]) # [left spacing, bottom spacing, width, height]
axis2.plot(x,y)
```

### Types of Plots

```python
plt.scatter(x,y)
```

```python
rand=np.random.randint(1,999, size=100)
plt.hist(rand)
```

```python
plt.boxplot(x)
```

### Saving Plot

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color="red")
plt.savefig("Demofig.png")
```
