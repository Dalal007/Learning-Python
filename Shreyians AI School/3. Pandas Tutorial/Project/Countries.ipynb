```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Countries.csv")
df.head()
```

```python
# which country has highest population
df[df["population"]==df["population"].max()]
```

```python
# which country has least population
df[df["population"]==df["population"].min()]
```

```python
# top 5 countries with highest democratic score
df.sort_values(by="democracy_score", ascending=False).head()
```

```python
# how many total regions are there
df.value_counts("region").count()
```

```python
# how many countries lies in Eastern Europe
df[df["region"]=="Eastern Europe"].shape
```

```python
# political leader of 2nd highest populated country
df[df["population"]==df["population"].nlargest(2).iloc[1]]
```

```python
# total countries whose political leaders are unknown
df[df["political_leader"].isna()]["country"].count()
```

```python
# total countries having Republic in their full name
count=0
def check_republic_country(name):
    global count
    if "republic" in name.lower():
        count+=1
    return count

df["country_long"].apply(check_republic_country)
print(count)
```

```python
# country in african region with highest population
df[df["population"]==df[df["continent"]=="Africa"]["population"].max()][["country","population"]]
```
