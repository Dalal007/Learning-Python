# Creating Dataframes

```python
data={
    'name':['John','Anna','Peter','Linda'],
    'age':[28,34,29,42],
    'city':['New York','Paris','Berlin','London'],
    'salary':[65000,70000,62000,85000]
}
df=pd.DataFrame(data)
df
```

```python
data_list=[
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2=pd.DataFrame(data_list)
custom_columns=["Name", "Age", "City", "Salary"]
df2=pd.DataFrame(data_list, columns=custom_columns)
df2
```

### Selection and indexing of columns

```python
df2["Name"]
```

```python
df2[["Name",'Salary']]
```

### Creating a new column

```python
df2["Designation"]=["Doctor","Engineer","Driver","Minister"]
df2
```

### Removing columns

```python
# By default axis=0(row). 'inplace' is used to make changes in actual data in memory
df2.drop("Designation",axis=1, inplace=True)
df2
```

```python
df2.drop([2,3],axis=0)
```

### Selecting Rows

```python
# use 'loc' to select rows
df2.loc[[1,2]]
```

```python
# use 'iloc' for indexed location
df2.iloc[[2,3]]
```

### Selecting subsets of Rows and Columns

```python
df
```

```python
df.loc[list(range(1, 4, 2)), ['city', 'salary']]
```

### Conditional Selection

```python
df2
```

```python
# Get only those who are above 30
df2[df2["Age"]>30]
```

```python
# Get only those who are above 30 and from Paris
df2[(df2["Age"]>30) & (df2["City"]=="Paris")]
```
