```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Matplotlib basaic

```python
x=np.linspace(1,10,11)
y=x**3
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.title("Number and its Cube")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

## Subplot in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,list(map(lambda x: x ** 4, x)))
```

## Object Oriented way

```python
fig=plt.figure(figsize=(12,8), dpi=125)
axis=fig.add_axes([0.1,0.1,0.8,0.8]) # [left,bottom,width,height]
axis.plot(x,y)

axis1=fig.add_axes([0.2,0.5,0.2,0.2]) # [left,bottom,width,height]
axis1.plot(x,y)
```

## Types of plots

```python
plt.scatter(x,y)
```

```python
from random import sample
data=sample(range(1,1000),100)
plt.hist(data)
```

## Saving plot

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
