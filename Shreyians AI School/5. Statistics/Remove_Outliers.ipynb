```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
arr=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
# Get quartile values
Q1=np.percentile(arr,25)
Q3=np.percentile(arr,75)
print("Q1=====>", Q1)
print("Q3=====>", Q3)
```

```python
IQR=Q3-Q1
print("IQR=====>", IQR)
```

```python
UF=Q3+1.5*IQR
LF=Q1-1.5*IQR
print("UF=====>", UF)
print("LF=====>", LF)
```

```python
# Remove outliers
l=[]
for i in arr:
    if i>=LF and i<=UF:
        l.append(i)

arr2=np.array(l)
print("Brfore=====>", arr)
print("After=====>", arr2)
```

```python
# Create 
plt.subplot(2,1,1)
sns.boxplot(x=arr) # Before
plt.subplot(2,1,2)
sns.boxplot(x=arr2) # After
```
