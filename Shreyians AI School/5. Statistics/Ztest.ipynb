```python
import numpy as np
from scipy.stats import norm
```

```python
sample=[172,174,168,169,171,173,175,170,169,172]
population_mean=170
population_std=3
sample_mean=np.mean(sample)
n=len(sample)
```

```python
z_score=(sample_mean-population_mean) / (population_std/np.sqrt(n))
z_score
```

```python
# If we take Alpha=0.05, our critical value will be -1.96 and +1.96. In this case we well ACCEPT our Null Hypothesis.
```

```python
p_value=2*(1-norm.cdf(abs(z_score))) # CDF= Cumulative Distribution Function
p_value
```

```python
alpha=0.05

if p_value < alpha:
    print("I will REJECT my Null Hypothesis")
else:
    print("I will ACCEPT my Null Hypothesis")
```
