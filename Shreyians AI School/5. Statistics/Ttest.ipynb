```python
import numpy as np
from scipy import stats
```

```python
sample=[172,174,168,169,171,173,175,170,169,172]
sample_mean=np.mean(sample)
sample_std=np.std(sample, ddof=1) # Just for an example we are taking ddof=1
n=len(sample)
mean_population=170
```

```python
t_stats=(sample_mean-mean_population) / (sample_std/np.sqrt(n))
t_stats
```

```python
p_value=2*(1-stats.t.cdf(abs(t_stats),df=n-1)) # CDF= Cumulative Distribution Function  # df= Degree of Freedom (sample_size-1).
p_value
```

```python
alpha=0.05

if p_value < alpha:
    print("I will REJECT my Null Hypothesis")
else:
    print("I will ACCEPT my Null Hypothesis")
```
